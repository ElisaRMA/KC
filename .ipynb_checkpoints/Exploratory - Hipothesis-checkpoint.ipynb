## 1.3 Data Transformations and Feature Engineering

```python
# transformation to datetime type
data['date'] = pd.to_datetime(data['date'], format = '%Y-%m-%d')
```

```python
# new columns of month and year
data['month'] = data['date'].dt.month
data['year'] = data['date'].dt.year
```

```python
# Season information

#     Spring: March 20, at 5:37 A.M.
#     Summer: June 20, at 11:32 P.M.
#     Fall: September 22, at 3:21 P.M.
#     Winter: December 21, at 10:59 A.M.
# https://www.calendarr.com/united-states/seasons-of-the-year-in-the-united-states/


data['season'] = data['month'].apply(lambda x: 'spring' if (x >=3) & (x<=5) else 
                                               'summer' if (x >=  6) & (x <= 8) else
                                               'fall' if (x >= 9 ) & (x <= 11) else
                                               'winter')
```

```python
# Was the house built before or after 1955? 
data['construction'] = data['yr_built'].apply(lambda data: '>1955' if data > 1955 else '<1955')
```

```python
# was the house renovated?
```

```python
# house condition
data['condition_binary'] = data['condition'].apply(lambda condition: 'good' if condition > 2 else 'bad') 
```

```python
# view quality
data['view_quality'] = data['view'].apply(lambda view: 'good' if view > 2 else 'bad')
```

```python
# does the house has a basement?
data['basement'] = data['sqft_basement'].apply(lambda sqft: 'yes' if sqft > 0 else 'no')
```

# Insights


## H1 - Waterfront houses are 30% more expensive.


```python
price_waterfront = data[['waterfront', 'price']].groupby('waterfront').mean('price').reset_index()
sns.barplot(y = price_waterfront['price'], x = price_waterfront['waterfront']);
```

## H2 - Houses built before 1955 are 50% less expensive.

```python
age_price = data[['construction', 'price']].groupby('construction').mean('price').reset_index()
sns.barplot(y = age_price['price'], x = age_price['construction']);
```

## H3 - Houses with a basement have 50% more sqft_lot than houses without basements.



```python
sqft_with_basement = data[['basement', 'sqft_lot']].groupby('basement').mean('sqft_lot').reset_index()
sns.barplot(y=sqft_with_basement['sqft_lot'], x=sqft_with_basement['basement']);
```

## H4 - The YoY rise in price is 10% 

```python
# only two years on date. Barplot is better
data.year.unique()
```

```python
yoy = data[['price', 'year']].groupby('year').sum('price').reset_index()
sns.barplot(y=yoy['price'],x=yoy['year']);
```

## H5 - The MoM rise in prices of houses with 3 bathroom is 15%

```python
mom = data[['month', 'price', 'bathrooms']].groupby(['month', 'bathrooms']).sum('price').reset_index()
mom_bath = mom[mom.bathrooms == 3] 

sns.lineplot(y=mom_bath.price, x= mom_bath.month, color = 'black', markers = True, style = (1), legend = False);
```

## H6 - The price of houses decrease 30% at every drop in condition (5 to 1)

```python
price_cond = data[['price', 'condition']].groupby('condition').mean('price').reset_index()
price_cond
```

```python
# sets the color pattern for the bars
colors = ['darkorange' if (x == 1 or x == 2) else 'grey' for x in price_cond['condition']]

# palette can be called in the barplot function
plot = sns.barplot(y= price_cond['price'],x=price_cond['condition'], palette = colors)
plot.bar_label(plot.containers[0]);
```

## H7 - Houses in bad condition (1-2) with good view (3 and 4) are 50% less expensive than houses in good condition and good view


```python
condition_view = data[['condition_binary', 'view', 'price']].groupby(['condition_binary', 'view']).mean('price').reset_index()
condition_view = condition_view.iloc[[1,3],[0,2]]
condition_view
```

```python
sns.barplot(x=condition_view.condition_binary ,y=condition_view.price);
```
